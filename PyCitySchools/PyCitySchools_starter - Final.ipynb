# PyCity Schools Analysis

# Charter Schools outperformed District schools in all categories of Avg Math Scores, Avg Reading Scores, % Passing Math, 
# % Passing Reading, and % Overall Passing.

# Medium and Smaller schools outperformed the largers in all categories, especially Math.

# As far as Spending Per Student, it appears the more dollars spent on students ironically reflected lower averages and lower passing percentages. 
# One surprising difference reflected a 27% variance between schools spending $585 or less, compared to those spending more($645 -$680). 

# Although the variances are somewhat surprising, further review would be required to understand overall dynamicos of each school to understand if there
# are underlying factors that need to be address.

```python
# Dependencies and Setup
import pandas as pd

from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)


school_data["budget"] = school_data["budget"].astype(int)


# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on="school_name")
school_data_complete.head()
```

## District Summary

```python
school_data_complete["school_name"].unique()
```

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete["school_name"].unique())

#school_count
school_count
```

```python
# Calculate the total number of students - count of student IDs

student_Count = school_data_complete["Student ID"].count()
#student_count- actual count of student IDs. Best not to use names as some may have the same name.

print(student_Count)
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
#total_budget
print(total_budget)
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
#average_math_score
print(average_math_score)
```

```python
# Calculate the average (mean) reading score
average_reading_score =school_data_complete["reading_score"].mean()
#average_reading_score
print(average_reading_score)
```

```python
# indentify the number of students that passed math in total
pass_Math_Students = school_data_complete[school_data_complete["math_score"] >= 70]
passing_math_count = len(pass_Math_Students)
passing_math_count
print(passing_math_count)
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)

passing_math_percentage = (passing_math_count / float(student_Count)) * 100
passing_math_percentage
print(passing_math_percentage)
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
pass_Reading_Students = school_data_complete[school_data_complete["reading_score"] >= 70]
passing_reading_count = len(pass_Reading_Students)
passing_reading_count
```

```python
# passing_reading_percentage =
passing_reading_percentage = (passing_reading_count / student_Count) * 100
passing_reading_percentage
print(passing_reading_percentage)
```

```python
# find all students in math and reading that scored 70 or higher
passingStudents = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
# passingStudents.head()
passing_math_reading_count = len(passingStudents)
passing_math_reading_count                       

```

```python
# Use the following to calculate the percentage of students that passed math and reading

overall_passing_rate = passing_math_reading_count /  float(student_Count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame - make a dictionary of lists
district_summary =pd.DataFrame(
    {
        "Total Schools": [school_count],
        "Total Students": [student_Count],
        "Total Budget": [total_budget],
        "Average Math Score": [average_math_score],
        "Average Reading Score": [average_reading_score],
        "% Passing Math": [passing_math_percentage],
        "% Passing Reading": [passing_reading_percentage],
        "% Overall Passing": [overall_passing_rate]
    }
)

# Formatting:
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("{:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.2f}%".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:.2f}%".format)

# Display the DataFrame
district_summary
```
